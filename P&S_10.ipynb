**MULTIPLE REGRESSION**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#Step 1: Create the dataset
data = {
'Size': [1000, 1200, 1500, 1800, 2200, 2500],
'Bedrooms': [2, 3, 3, 4, 4, 4],
'Price': [150, 180, 210, 250, 300, 360]
#in square feet
#number of bedrooms
#in $10005
}
df = pd.DataFrame(data)
#Step 2: Define features (X) and target (y)
x = df[['Size', 'Bedrooms']]
#multiple predictors
y=df['Price']
#target variable
x = df[['Size', 'Bedrooms']]
#multiple predictors
y=df['Price']
#target variable
```

```python
#Step 3: Create and train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y) # Changed 'X' to 'x'
#Step 4: Predict prices using the trained model
df[ 'Predicted_Price'] = model.predict(x) # Changed 'X' to 'x'
#Step 5: Print coefficients and intercept
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
print("\nPrediction Table:")
print(df)
```

```python
# Step 6: Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.scatter(df['Size'], df ['Price'], color='blue', label='Actual Price')
plt.plot(df['Size'], df['Predicted_Price'], color='red', label='Predicted Price (Linear)')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($1000s)')
plt.title('Multiple Linear Regression: House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```

**POLYNOMIAL REGRESSION**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
#Step 1: Create the dataset
data = {
'Speed': [10, 20, 30, 40, 50, 60, 70, 80], # in km/h
'Distance': [5, 9, 15, 25, 38, 55, 75, 98] # in meters
}
df = pd.DataFrame(data)
#Step 2: Prepare features
X = df [['Speed']]
# Use 'Distance' from the current dataframe for y
y = df['Distance'] # This is the change

# Step 3: Transform features to polynomial (degree 2)
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
# Step 4: Fit the polynomial regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_poly, y)
# Step 5: Make predictions
df['Predicted'] = model.predict(X_poly)
```

```python
#Step 6: Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(df['Speed'], df['Distance'], color='blue', label='Actual Data')
plt.plot(df['Speed'], df ['Predicted'], color='red', label='Polynomial Fit')
plt.xlabel('Speed (km/h)')
plt.ylabel('Stopping Distance (m)')
plt.title('Polynomial Regression: Speed vs Stopping Distance')
plt.legend()
plt.grid(True)
plt.show()
#Step 7: Print model coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
