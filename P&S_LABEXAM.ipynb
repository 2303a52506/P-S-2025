```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
N=1000
data = np.random.choice([0,1,2,3,4,5,6],size=N, p=[0.1, 0.25, 0.25, 0.1, 0.05, 0.05,0.2])
unique,counts = np.unique(data,return_counts=True)
pmf=counts/N
cdf=np.cumsum(pmf)
plt.figure(figsize=(10,5))
sns.barplot(x=unique,y=pmf,color="skyblue",edgecolor="black")
plt.xlabel("Number of children")
plt.ylabel("Probability")
plt.title("Probability of number of children")
plt.xticks(unique)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=unique,y=cdf,color='blue',linestyle='--')
plt.xlabel("Number of children")
plt.ylabel("Probability")
plt.title("Probability of number of children")
plt.xticks(unique)
plt.ylim(0,1.1)
plt.show()
```
