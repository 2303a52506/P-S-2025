```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/content/employee_data_250.csv'
employee_data_250 = pd.read_csv(file_path)
print(employee_data_250.head())
```

```
plt.figure(figsize=(10, 6))
plt.hist(employee_data_250['Salary'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Salary', fontsize=16)
plt.xlabel('Salary (USD)', fontsize=14)
plt.ylabel('Number of Employees', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
plt.figure(figsize=(10, 6))
employee_counts = employee_data_250['Department'].value_counts()
plt.bar(employee_counts.index, employee_counts.values, color='lightcoral', edgecolor='black')
plt.title('Number of Employees in Each Department', fontsize=16)
plt.xlabel('Department', fontsize=14)
plt.ylabel('Number of Employees', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
plt.figure(figsize=(12, 8))
job_level_department = employee_data_250.groupby(['Department', 'Job Satisfaction']).size().unstack(fill_value=0)
job_level_department.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='coolwarm', edgecolor='black')
plt.title('Job Level Distribution Across Departments', fontsize=16)
plt.xlabel('Department', fontsize=14)
plt.ylabel('Number of Employees', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Job Satisfaction', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```
salary_range = pd.cut(employee_data_250['Salary'], bins=10).value_counts().idxmax()
print(f"The most common salary range is: {salary_range}")
```

```
correlation = employee_data_250['Experience (Years)'].corr(employee_data_250['Salary'])
print(f"The correlation between Experience and Salary is: {correlation:.2f}")
```

```
highest_department = employee_counts.idxmax()
print(f"The department with the highest number of employees is: {highest_department}")
```
